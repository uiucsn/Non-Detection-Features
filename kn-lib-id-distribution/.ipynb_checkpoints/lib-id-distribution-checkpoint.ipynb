## Sky distribution of LIBIDs that were inside 90% confidence regions in a skymap

```python
from astropy import coordinates, units as u
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_skymap_libid_mapping = pd.read_pickle('lib_id_skymap_ninety_map.pickle')
df_lib_id_location = pd.read_pickle('ztf_dr3_lib_id_map.pickle')
```

- The skymaps corresponding to a found KN from the Bulla model.
- The LIB_IDs are those within the 90% confidence region of each skymap

```python
df_skymap_libid_mapping.head()
```

- Sky location of LIB_IDs
- Count corresponds to the number of times they were inside 90% localization of a skymap

```python
df_lib_id_location.head()
```

```python
ra = coordinates.Angle(df_lib_id_location.ra.values, unit=u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coordinates.Angle(df_lib_id_location.dec.values, unit=u.degree)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="mollweide")

c = ax.scatter(
    ra.radian, dec.radian,
    c=df_lib_id_location.counts.values,
    s=2, cmap='viridis_r',
)
plt.colorbar(c, label='Count', orientation="horizontal")
ax.grid()
```
